```python
def problem(dto, jerk, s1, s2, s3, s4, s5, s6):
    return 2.81*dto + 0.14*jerk + 4.63*s1 + 7.34*s2 + 10.09*s3 + 12.9*s4 + 15.63*s5 + 17.58*s6 - 1

def fit(dto, jerk, s1, s2, s3, s4, s5, s6):
    ans = problem(dto, jerk, s1, s2, s3, s4, s5, s6)
    if ans == 0:
        return 99999
    else:
        return abs(1/ans)
```

```python
class GeneticAlgorithm():
    def __init__(self, startingPopulation=10000, childredPopulation=1000, maxGenerations=10000, numberOfVariables=8, sd=0.1, mutationChance=0.4):
        self.startingPopulation = startingPopulation
        self.childredPopulation = childredPopulation
        self.maxGenerations = maxGenerations
        self.numberOfVariables = numberOfVariables
        self.solutions = [tuple(random.uniform(0, 10000) for _ in range(self.numberOfVariables)) for _ in range(self.startingPopulation)]
        self.sd = sd
        self.mutationChance = mutationChance
    
    
    def problem(self, dto, jerk, s1, s2, s3, s4, s5, s6):
        return 2.81*dto + 0.14*jerk + 4.63*s1 + 7.34*s2 + 10.09*s3 + 12.9*s4 + 15.63*s5 + 17.58*s6 - 1


    def fitness(self, dto, jerk, s1, s2, s3, s4, s5, s6):
        ans = self.problem(dto, jerk, s1, s2, s3, s4, s5, s6)
        if ans == 0:
            return 99999
        else:
            return abs(1/ans)


    def evaluate(self):
        newGen = None
        rankedSolutions = None
        for i in range(self.maxGenerations):
            rankedSolutions = []
            for s in self.solutions:
                rankedSolutions.append((self.fitness(s[0], s[1], s[2], s[3], s[4], s[5], s[6], s[7]), s )) # fitness value, parameters
            rankedSolutions.sort(reverse=True)

            if i % 100 == 0:
                print(f"=== Gen {i} best solution ===", end="\n\t")
                print(f"Fitness: {str(round(rankedSolutions[0][0], 8)).ljust(15)} Values: {rankedSolutions[0][1]}")
                # print(f"=== Gen {i} best solution === Fitness: {str(round(rankedSolutions[0][0], 8)).ljust(15)} Values: {rankedSolutions[0][1]}", end="\r")

            if rankedSolutions[0][0] > 9999: # exit if the fitness is good enough
                print(f"\n=== Fit enough in gen {i} ===", end="\n\t")
                print(f"Fitness: {str(round(rankedSolutions[0][0], 8)).ljust(15)} Values: {rankedSolutions[0][1]}")
                res = self.problem(rankedSolutions[0][1][0], rankedSolutions[0][1][1], 
                            rankedSolutions[0][1][2], rankedSolutions[0][1][3], 
                            rankedSolutions[0][1][4], rankedSolutions[0][1][5], 
                            rankedSolutions[0][1][6], rankedSolutions[0][1][7]) + 1
                # print(f"Test: {res}")
                return rankedSolutions[0]

            newGen = [solution[1] for solution in rankedSolutions[:10]] # Keep n best from previous generation

            for _ in range(self.childredPopulation): # The new generation consists of 10 + self.childredPopulation solutions
                parent1, parent2 = random.sample(rankedSolutions[:100], 2)
                children = self._crossover(parent1[1], parent2[1])
                for child in children:
                    if random.random() < self.mutationChance:
                        newGen.append(self._mutation(child))
                    else:
                        newGen.append(child)
                
            # # only 10 best solutions gets to have children with someone else
            # for parent in rankedSolutions[:10]:
            #     for _ in range(self.childredPopulation//10):
            #         while True:
            #             parent2 = random.sample(rankedSolutions[:100], 1)
            #             if parent != parent2:
            #                 break
            #         # print(parent, parent2)
            #         children = self._crossover(parent[1], parent2[0][1])
            #         # print("Children:")
            #         # print(children)
            #         # return
            #         for child in children:
            #             if random.random() < self.mutationChance:
            #                 newGen.append(self._mutation(child))
            #             else:
            #                 newGen.append(child)
                # print(f"Family: \n\t{parent1[1]}\n\t{parent2[1]}\n\t{newGen[-1]}\n\t{newGen[-2]}")
            self.solutions = newGen
        else:
            # Only prints this if the results are not fit enough
            print(f"\n=== Best solution after {self.maxGenerations} generations ===", end="\n\t")
            print(rankedSolutions[0])
            res = self.problem(rankedSolutions[0][1][0], rankedSolutions[0][1][1], 
                        rankedSolutions[0][1][2], rankedSolutions[0][1][3], 
                        rankedSolutions[0][1][4], rankedSolutions[0][1][5], 
                        rankedSolutions[0][1][6], rankedSolutions[0][1][7]) + 1
            print(f"Test: {res}")

    def _crossover(self, parent1, parent2):
        child1 = []
        child2 = []
        for g1, g2 in zip(parent1, parent2):
            if random.randint(0, 1):
                child1.append(g1)
                child2.append(g2)
            else:
                child1.append(g2)
                child2.append(g1)
        return tuple(child1), tuple(child2)


    def _mutation(self, solution):
        sol = list(solution)
        for i in range(len(sol)):
            if random.randint(0, 1):
                sol[i] *= random.uniform(1-self.sd, 1+self.sd)
        return tuple(sol)
    

    def run(self):
        return self.evaluate()
```
